## null imputation

```python
# 1. If this is a numerical column, then we can impute a null value with the column mean or median
# 2. If the value is a categorical (text), we can impute the null value with the column mode.
# 3. We can impute the null values with value before the null or after the null.
# 4. Domain specific imputation
```

```python
test_data["HP"].head(10)
```

### fillna Function

```python
# null value update in a cell in a DataFrame
test_data.loc[1,"HP"]= None
test_data
```

```python
# mean value output in "HP" column in DataFrame(test_data)
hp_mean = test_data["HP"].mean() # the mean value stores in a variable
hp_mean
```

```python
# mean imputation in null value in "HP" column in DataFrame(test_data)
test_data["HP"].fillna(hp_mean,inplace = True) # here we also permanent update the null value
test_data
```

```python
# null value update in a cell in a DataFrame
test_data.loc[2,"Generation"]= None
test_data
```

```python
# median value output in "Generation" column in DataFrame(test_data)
hp_median = test_data["Generation"].median() # the median value stores in a variable
hp_median
```

```python
# median imputation in null value in "Generation" column in DataFrame(test_data)
test_data["Generation"].fillna(hp_median,inplace = True) # here we also permanent update the null value
test_data
```

```python
# null value update in a cell in a DataFrame
test_data.loc[0,"Type 1"]= None
test_data
```

```python
# mode value output in "Type 1" column in DataFrame(test_data)
hp_mode = test_data["Type 1"].mode()[0] # the mode value stores in a variable
hp_mode
```

```python
# mode imputation in null value in "Type 1" column in DataFrame(test_data)
test_data["Type 1"].fillna(hp_mode,inplace = True) # here we also permanent update the null value
test_data
```

```python
# user defined value imputation in null values
test_data["Type 2"].fillna("Deepjoy").head(20)
```

```python
# impute the null values with value before the null (forward filling) in "Type 2" column in DataFrame(test_data)
test_data["Type 2"].fillna(method="ffill").head(20)
```

```python
test_data["Type 2"].head(20)
```

```python
# impute the null values with value after the null (backward filling) in "Type 2" column in DataFrame(test_data)
test_data["Type 2"].fillna(method="bfill").head(20)
```
