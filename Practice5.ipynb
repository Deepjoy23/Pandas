```python
#First Letter of each cell value of a Column("Company1") of a DataFrame(carPriceDf) and stored a new column (processing1)
def fetch_1st_letter(x):
    return x[0]

carPriceDf["processing1"]=carPriceDf["Company1"].apply(fetch_1st_letter)
carPriceDf
```

```python
#First two Letters of each cell value of a Column("Company1") of a DataFrame(carPriceDf) and stored a new column (processing1)
def fetch_two_letters(x):
    return x[:2]

carPriceDf["processing2"]=carPriceDf["Company1"].apply(fetch_two_letters)
carPriceDf
```

```python
#First last Letter of each cell value of a Column("Company1") of a DataFrame(carPriceDf) and stored a new column (processing3)
def fetch_last_letter(x):
    return x[-1]

carPriceDf["processing3"]=carPriceDf["Company1"].apply(fetch_last_letter)
carPriceDf
```

```python
#csv type file access by pandas
test_data=pd.read_csv("pokemon_data.csv")
test_data
```

```python
# speed_category --> >=100 --> high speed, >=80 --> medium speed, >=60 --> low speed, slow
def speed_category(x):
    if x>=100:
        return "high speed"
    elif x>=80:
        return "medium speed"
    elif x>=60:
        return "low speed"
    else:
        return "slow"

test_data["Speed Checking"]=test_data["Speed"].apply(speed_category)
test_data.head(10)
```

```python
# speed_category --> >=100 --> high speed 10 deepjoy, >=80 --> medium speed 10 deepjoy, >=60 --> low speed 10 deepjoy, slow 10 deepjoy
def speed_category(x,num,text):
    if x>=100:
        return f"high speed {num} {text}"
    elif x>=80:
        return f"medium speed {num} {text}"
    elif x>=60:
        return f"low speed {num} {text}"
    else:
        return f"slow {num} {text}"

test_data["Speed Checking"]=test_data["Speed"].apply(speed_category,args=(10,"deepjoy"))
test_data.head(10)  
```

```python
# speed_category --> >=100 --> high speed, >=80 --> medium speed, >=60 --> low speed, slow
def speed_category(x):
    if x>=100:
        return "high speed"
    elif x>=80:
        return "medium speed"
    elif x>=60:
        return "low speed"
    else:
        return "slow"
# using for loop and list
list_ = []
for x in test_data['Speed']:
    value = speed_category(x)
    list_.append(value)
list_ 
```

```python
# Result printing
test_data["Speed Category"] = list_
test_data
```
