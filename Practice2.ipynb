### New Column Addition

```python
#Copy a DataFrame variable(data) in a DataFrame another variable(test_data)
test_data=data.copy()
test_data
```

```python
test_data["test"]=5
test_data
```

```python
test_data["Total"]= test_data["HP"]+test_data["Attack"]+test_data["Defense"]
test_data.head()
```

### Column Deletation

```python
# droping multiple columns but not update
test_data.drop(["test","Total"],axis=1)
test_data.head()
```

```python
# droping multiple columns and update
test_data=test_data.drop(["test","Total"],axis=1)
test_data.head()
```

```python
# droping multiple columns and update
test_data.drop(["test","Total"],axis=1,inplace=True) # update at a glance
test_data.head()
```

```python
test_data.head()
```

```python
# droping multiple rows but not update
test_data.drop([0,3],axis=0)
test_data.head()
```

```python
# droping multiple rows and update
test_data=test_data.drop([0,3],axis=0)
test_data.head()
```

```python
# droping multiple rows and update
test_data.drop([0,3],axis=0,inplace=True)
test_data.head()
```
