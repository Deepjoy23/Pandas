## group by

```python
# give the avgerage Attack for Type 1 in the DataFrame(test_data)
test_data.groupby("Type 1")["Attack"].mean() # in a Boolean output(Serie Formant, (index(1)=Type 1) and values= mean)
                                             # 1st hierarchy index is : Type 1
```

```python
# give the avgerage Attack for Type 1
test_data.groupby("Type 1")["Attack"].mean().reset_index() # in a DataFrame output
```

```python
# give the avgerage Attack for Type 1 and Legendary
test_data.groupby(["Type 1","Legendary"])["Attack"].mean() # in a Boolean output (Serie Formant, (index(2)=Type 1,Legendary) and values= mean)
                                                           # 1st hierarchy index is : Type 1 and 2nd hierarchy index is : Legendary
```

```python
# give the avgerage Attack for Type 1 and Legendary
test_data.groupby(["Type 1","Legendary"])["Attack"].mean().reset_index() # in a DataFrame output
```

```python
# DataFrame
test_data
```

```python
# give the avgerage (Attack Defense, Sp. Atk, Sp. Def, Speed) for Type 1 and Legendary
test_data.groupby(["Type 1","Legendary"])[["Attack","Defense","Sp. Atk","Sp. Def","Speed"]].mean() # Normal Output without index format
```

```python
# give the avgerage (Attack Defense, Sp. Atk, Sp. Def, Speed) for Type 1 and Legendary
test_data.groupby(["Type 1","Legendary"])[["Attack","Defense","Sp. Atk","Sp. Def","Speed"]].mean().reset_index() # in a DataFrame output
```

## concat

```python
a = {"Roll":[1,2,3],"Name": ["Ram Sarkar","Rahim Mondal","Tina Basak"], "Marks":[50,45,63]}
b = {"Roll":[4,5,6],"Name":["Ram Hazra","Rahim Das","Neela Dutta"], "Marks":[52,44,69]}
df1 = pd.DataFrame(a)
df2 = pd.DataFrame(b)
```

```python
df1
```

```python
df2
```

```python
# horizontal concatination of two DataFrame
pd.concat([df1,df2],axis=1)
```

```python
# vertical concatination of two DataFrame
pd.concat([df1,df2]) # default vertical concatination
```

```python
# vertical concatination of two DataFrame
pd.concat([df1,df2]).reset_index() # index wise outout
```

```python
# vertical concatination of two DataFrame
pd.concat([df1,df2]).reset_index(drop=True) # index droping
```

```python
# vertical concatination of two DataFrame
deepjoy=pd.concat([df1,df2]).reset_index(drop=True) # index droping
deepjoy
```

## join/merge

```python
# for same column names joining/merging process
Car_Price = {'Company': ['Toyota', 'BMV', 'Honda', 'Audi','Tata'], 'Price': [23845, 17995, 135925 , 71400, 45000]}
carPriceDf = pd.DataFrame.from_dict(Car_Price)

car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi','Maruti'], 'horsepower': [141, 80, 182 , 160, 50]}
carsHorsepowerDf = pd.DataFrame.from_dict(car_Horsepower)
```

```python
carPriceDf
```

```python
carsHorsepowerDf
```

```python
carPriceDf.merge(carsHorsepowerDf,on="Company") # default inner join
```

```python
# inner join
carPriceDf.merge(carsHorsepowerDf,on="Company",how="inner")
```

```python
# outer join
carPriceDf.merge(carsHorsepowerDf,on="Company",how="outer")
```

```python
# right join
carPriceDf.merge(carsHorsepowerDf,on="Company",how="right")
```

```python
# left join
carPriceDf.merge(carsHorsepowerDf,on="Company",how="left")
```

```python
# for diffrent column names joining/merging process
Car_Price = {'Company1': ['Toyota', 'BMV', 'Honda', 'Audi','Tata'], 'Price': [23845, 17995, 135925 , 71400, 45000]}
carPriceDf = pd.DataFrame.from_dict(Car_Price)

car_Horsepower = {'Company2': ['Toyota', 'Honda', 'BMV', 'Audi','Maruti'], 'horsepower': [141, 80, 182 , 160, 50]}
carsHorsepowerDf = pd.DataFrame.from_dict(car_Horsepower)
```

```python
carPriceDf
```

```python
carsHorsepowerDf
```

```python
# two different column_names of two different dataFrame inner join 
carPriceDf.merge(carsHorsepowerDf,how="inner",left_on="Company1",right_on="Company2")
```
